### Reshaping the data

```python
census_subset = census[['CensusId', 'State', 'County', 'TotalPop', 'Men', 'Women', 'Hispanic', 'Asian', 'Black', 'Native', 'Pacific', 'White']]
```

```python
census_subset.head()
```

## Melting

```python
census_reshaped = census_subset.melt(id_vars=['CensusId', 'TotalPop', 'State', 'County','Men', 'Women'], var_name = 'Race', value_name = 'RacePercent')
census_reshaped.head()
```

```python
census_reshaped['RacePop'] = census_reshaped['RacePercent'] * census_reshaped['TotalPop'] / 100
census_reshaped.head()
```

## Exploratory plots

### Ratial proportions

```python
plt.bar(census_reshaped['Race'], census_reshaped['RacePop'])
plt.show()
```
