```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline

pd.set_option("max.columns", 1000)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies.describe()
```

```python
movies['release_date'] = movies['release_date'].astype(str)
```

```python
def ExtractYear(val):
    tokens = val.split('-')
    return tokens[0]
```

```python
movies['release_year'] = movies['release_date'].apply(ExtractYear)
```

```python
movies.head()
```

```python
movies['release_year'] = pd.to_numeric(movies['release_year'], errors='coerce')
```

## Breaking out the genres

```python
import ast

def ExtractGenreAsList(dictionary_genre):
    dictionary_genre = ast.literal_eval(dictionary_genre)
    list_genres = []
    for entry in dictionary_genre:
        list_genres.append(entry['name'])
    return list_genres
```

```python
movies["genre_list"] = movies["genres"].apply(ExtractGenreAsList)
movies[["genres","genre_list"]].head()
```

```python
movies["genre_list"].head()
```

## Extract unique genres

```python
movies.head()
```

```python
unique_genres = []
for row in movies["genre_list"]:
    for genres in row:
        unique_genres.append(genres)

#print(unique_genres)

unique_genres = set(unique_genres)

print(unique_genres)
```

```python
def GetGenrePerRow(row, genre):
    if genre in row:
        return 1
    else:
        return 0
    
```

```python
for genre in unique_genres:
    movies[genre] = movies["genre_list"].apply(lambda x: GetGenrePerRow(x, genre))
```
